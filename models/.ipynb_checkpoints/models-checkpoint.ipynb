```python
import pandas as pd
df = pd.read_csv('./data/airbnb_data_cleaned.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
train_data = df[df['property_size'].notnull()]
test_data  = df[df['property_size'].isnull()]

# define X
X_train = train_data.drop('property_size', axis=1)
X_test  = test_data.drop('property_size', axis=1)

# define y
y_train = train_data['property_size']

# import Linear Regression
from sklearn.linear_model import LinearRegression

# instantiate
linreg = LinearRegression()

# fit model to training data
linreg.fit(X_train, y_train)

y_test = linreg.predict(X_test)
```
